Here is an example function that will help get data out of the .txt files I have provided.

```python
def get_fw_data(file,code):
    
    # file is the file we want to read from
    # code is the IZA framework code you want data on
    
    file = open(file,'r')
    data = file.readlines()
    file.close()
    
    for i,line in enumerate(data):
        fields = line.split()
        if fields[0] == code:
            start = i
            break
    
    # go through fw chunk and get data
    
    t_sites = {}
    o_sites = {}
    for i,line in enumerate(data[start+1:]):
        line = line.rstrip(' \n')

        # get the t site rings
        
        if 'T' in line and ':' in line:
            fields = line.split(':')
            t_sites[fields[0]]=fields[1]                
        
        # get the o site rings
        
        if 'O' in line and ':' in line:
            fields = line.split(':')
            o_sites[fields[0]]=fields[1]        
        
        fields = line.split()
        if fields[0] == 'Framework':
            end = i+start-1
            break

    return(t_sites,o_sites)    
#nineteen columns vector of all zeros and each column represents a number of tsites and osites (separate vectors)
#save as a panda data table
```

```python
def nringsvector(code,file):
    '''takes in the file and code name and returns how many of each ring type is present
    arguments: file: dataset
                code: which zeolite is being observed
    returns: vector of number of rings for each tsite'''
    
    #get site data
    tsites, osites = get_fw_data(file,code)
    
    #create empty vector
    A = np.zeros([len(tsites),21])
    
    #coutner for which row of the matrix you are in
    k=0 
    
    #go through each tsite
    for t in tsites:
        #initialize accumulator 
        j=1
        
        #make list of number of rings
        tsites_list = tsites[t].split('_')
        #go through the string for each t list
        for i in range(0,len(tsites_list)):
            #make the while loop true each for loop
            yes = True
            while yes:
                if int(tsites_list[i]) == j:
                    A[k][j] = A[k][j] + 1
                    #end while loop 
                    yes = False
                #increase j 
                else:
                    j = j + 1
        print('1')
        A[0,k] = t
        #increase row
        k = k +1 
    return A
```

```python
def assemblematrix(file):
    fws = get_all_fws()
    fws = fws[1:]
    B = nringsvector('ABW',file)
    for code in fws:
        print(code)
        C = nringsvector(code,file)
        B = np.append(B, C, axis=0)
    return B
```

```python
#come up with the fws column
def fws_column(file):
    fws = get_all_fws()

    fws_times_tsites = []
    for code in fws:
        tsites, osites = get_fw_data(file,code)
        for t in tsites:
            fws_times_tsites.append(code)
    return fws_times_tsites
```

```python
def tsite_column(file):
    fws = get_all_fws()
    
    tsite_col = []
    for code in fws:
        tsites, osites = get_fw_data(file,code)
        for t in tsites:
            tsite_col.append(t)
    return tsite_col
```

```python
file = '../Data/crum_rings.txt'
framework = fws_column(file)
tsite = tsite_column(file)
value_matrix = assemblematrix('../Data/crum_rings.txt')
```
