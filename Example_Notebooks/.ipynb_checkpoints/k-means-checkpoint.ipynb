```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import GridSearchCV, learning_curve, train_test_split
from sklearn.utils.random import sample_without_replacement
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn.metrics import silhouette_samples, silhouette_score, completeness_score, homogeneity_score, mean_squared_error
from sklearn.cluster import KMeans
import pickle
import pylab
import warnings; warnings.simplefilter('ignore')
from yellowbrick.cluster import KElbowVisualizer, SilhouetteVisualizer
from sklearn.mixture import GaussianMixture
from sklearn import decomposition
from sklearn import random_projection
from yellowbrick.features import Manifold
```

```python
df = pd.read_csv('Data/phpbL6t4U.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,:-1], df.iloc[:,-1], test_size=0.2, random_state=0)
```

```python
scaler = MinMaxScaler()
scaler.fit(X_train, y_train)
```

```python
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
seed = 20210324
```

# Elbow method

```python
# Instantiate the clustering model and visualizer
model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,11))
visualizer.fit(X_train_scaled)        # Fit the data to the visualizer
# visualizer.show(outpath="Figures/elbow.pdf")        # Finalize and render the figure
```

# Silhouette visualizer

```python
model = KMeans(2, random_state=seed)
visualizer = SilhouetteVisualizer(model, colors='yellowbrick')
visualizer.fit(X_train_scaled)        # Fit the data to the visualizer
# visualizer.show(outpath="Figures/Silhouett_coefficient.pdf")        # Finalize and render the figure
```

# Visualizing high-dimensional data on a 2D plot

```python
classes = ['0','1','2','3','4','5']
# Instantiate the visualizer
viz = Manifold(manifold="tsne",classes = classes)
km = KMeans(n_clusters = 6, random_state = seed)
km.fit(X_train_scaled)
km_labels = km.predict(X_train_scaled)
viz.fit_transform(X_train_scaled,km_labels)  # Fit the data to the visualizer
# viz.show(outpath="Figures/tsne.pdf") 
```

# Visualize the clusters vs the real classes. There are five classes and two clusters.

```python
km = KMeans(n_clusters = 2, random_state = seed)
km.fit(X_train_scaled)
km_labels = km.predict(X_train_scaled)
cluster_labels = km.fit_predict(X_train_scaled, y=y_train)
```

```python
tmp = pd.DataFrame()
tmp['y'] = y_train
tmp['cluster'] = cluster_labels
df = pd.DataFrame(tmp.groupby(['y', 'cluster']).size()).reset_index()
df.columns = ['y', 'cluster', 'num']
df = pd.pivot_table(df,index=['y'],columns='cluster',values='num',aggfunc='sum').reset_index()
df.drop('y',axis=1,inplace=True)
```

```python
df.plot.bar(stacked=True)
plt.xticks(rotation = 0)
plt.xlabel('Train Label')
plt.ylabel('Cluster')
# plt.savefig("Figures/cluster_label_1.pdf",dpi=300)
plt.show()
```

```python
tmp = pd.DataFrame()
tmp['y'] = y_train
tmp['label'] = cluster_labels
df = pd.DataFrame(tmp.groupby(['y', 'label']).size()).reset_index()
df.columns = ['y', 'label', 'num']
df = pd.pivot_table(df,index=['label'],columns='y',values='num',aggfunc='sum').reset_index()
df.drop('label',axis=1,inplace=True)
```

```python
df.plot.bar(stacked=True)
plt.xticks(rotation = 0)
plt.ylabel('Train Label')
plt.xlabel('Cluster')
# plt.savefig("Figures/cluster_label_2.pdf",dpi=300)
plt.show()
```
