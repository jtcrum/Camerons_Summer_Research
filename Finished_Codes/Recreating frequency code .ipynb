```python
import matplotlib.pyplot as plt

import numpy as np

import pandas as pd

from zse.collections import *
from zse.utilities import *

from ase.io import read, write
from matplotlib import cm

from collections import Counter
```

```python
def get_fw_data(file,code):
    
    # file is the file we want to read from
    # code is the IZA framework code you want data on
    
    file = open(file,'r')
    data = file.readlines()
    file.close()
    
    for i,line in enumerate(data):
        fields = line.split()
        if fields[0] == code:
            start = i
            break
    
    # go through fw chunk and get data
    
    t_sites = {}
    o_sites = {}
    for i,line in enumerate(data[start+1:]):
        line = line.rstrip(' \n')

        # get the t site rings
        
        if 'T' in line and ':' in line:
            fields = line.split(':')
            t_sites[fields[0]]=fields[1]                
        
        # get the o site rings
        
        if 'O' in line and ':' in line:
            fields = line.split(':')
            o_sites[fields[0]]=fields[1]        
        
        fields = line.split()
        if fields[0] == 'Framework':
            end = i+start-1
            break

    return(t_sites,o_sites)    
```

```python
###goetzke 
#get all the data for goetzske

fws = get_all_fws()
#vector stores the number of frameworks that have each sized ring
goetzke_vector = np.zeros(21)
#go through ever framework
for code in fws:
    #get indiivudal tsite framework
    tsites,osites = get_fw_data('../Data/goetzke_rings.txt',code)
    #go through tsites 
    tsites_list = []
    for t in tsites:
        #print(t)
        #make it a list of integers
        tsites_list.extend(tsites[t].split('_')) 
    #go through and eliminate repeat values
    tsites_list = [ int(x) for x in tsites_list ]
    tsites_list.sort()
    n = int(tsites_list[0])
        
    goetzke_vector[n] = goetzke_vector[n] + 1
        
        
    #run through list
    for i in range(len(tsites_list)):
        if n != int(tsites_list[i]):
            n = int(tsites_list[i])
            goetzke_vector[n] = goetzke_vector[n] + 1
print(goetzke_vector)

```
